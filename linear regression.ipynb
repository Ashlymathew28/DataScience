```python
x_val=pd.DataFrame(data['x'])
x_median=x_val.median()
```

```python
# IQR of age
q3=x_val.quantile(q=.75)
q1=x_val.quantile(q=.25)
iqr=q3-q1

# deriving bounderies of outlier
iqr_ll=int(q1 - 1.5 * iqr)
iqr_ul=int(q3 + 1.5 * iqr)

#finding and treating outliers lower limit and upper limit
data.loc[data['x']>iqr_ul,'x'] = int(x_val.quantile(q=.99))
data.loc[data['x']<iqr_ll,'x'] = int(x_val.quantile(q=.01))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
regression=LinearRegression()
y_pred=regression.fit(x_train,y_train)
```

```python
accuracy=regression.score(x_test,y_test)
print("acccuracy",accuracy)
```

```python
test=pd.read_csv("C:/Users/ashly/OneDrive/Desktop/DataScience Anaconda/Regression/test.csv")
```

```python
test_x=test.iloc[:,:-1]
test_y=test.iloc[:,1]
y_pred=regression.predict(test_x)
y_pred
```

```python
test_y
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
import numpy as np
mae=mean_absolute_error(test_y,y_pred)
mse=mean_squared_error(test_y,y_pred)
rmse=np.sqrt(mse)
r2=r2_score(test_y,y_pred)
```

```python
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"R-squared (R2): {r2:.2f}")
```
