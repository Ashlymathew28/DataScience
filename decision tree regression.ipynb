```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

data=pd.read_csv("C:/Users/ashly/OneDrive/Desktop/DataScience Anaconda/Regression/insurance.csv")
pd.set_option('display.max_column',None)
pd.set_option('display.max_rows',None)
data
```

```python
data['sex']=data['sex'].astype('category')
data['sex']=data['sex'].cat.codes
```

```python
data
```

```python
data['smoker']=data['smoker'].astype('category')
data['smoker']=data['smoker'].cat.codes

data['region']=data['region'].astype('category')
data['region']=data['region'].cat.codes
```

```python
data.isnull().sum()
```

```python
x=data.drop(columns='charges')
y=data['charges']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
reg=DecisionTreeRegressor()
reg.fit(x_train,y_train)
```

```python
y_pred_train=reg.predict(x_test)
y_pred_train
```

```python
from sklearn.metrics import mean_squared_error,r2_score
mse=mean_squared_error(y_test,y_pred_train)
mse
```

```python
r2=r2_score(y_test,y_pred_train)
```
